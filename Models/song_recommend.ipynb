```python
import pandas as pd
import numpy as np
```

```python
data.drop_duplicates(inplace=True , subset=['name'])
data
```

```python
name = data['name']
name
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
col_features = ['danceability', 'energy', 'valence', 'loudness' , 'acousticness', 'instrumentalness' , 'liveness' , 'speechiness' ,'tempo']
X = MinMaxScaler().fit_transform(data[col_features])
kmeans = KMeans(init="k-means++",
                n_clusters=3,
                random_state=42)
kmeans.fit(X)
data['kmeans'] = kmeans.labels_
```

```python
data['song_name']=name
data
```

```python
cluster=data.groupby(by=data['kmeans'])
cluster
```

```python
y = data['kmeans']
x=data.drop(columns=['name','artists','id','release_date','song_name'])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
!pip install lightgbm
from lightgbm import LGBMClassifier
model=LGBMClassifier().fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
import lightgbm
import matplotlib.pyplot as plt
ax = lightgbm.plot_importance(model, max_num_features=10, figsize=(15,15))
plt.show()
```

```python
df=cluster.apply(lambda x: x.sort_values(["popularity"],ascending=False))
# df.reset_index(level=0, inplace=True)

df
```

```python
def get_results(emotion):
  
  if emotion=="happy":
      return df[df['kmeans']==0]['song_name'].head(100)

  elif emotion=="sad":
      return df[df['kmeans']==1]['song_name'].head(100)
      
  else:
      return(df[df['kmeans']==2]['song_name'].head(100)

```
